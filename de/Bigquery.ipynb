```python
from google.cloud import bigquery

# 1. Initialize the client
client = bigquery.Client()

# 2. Write your SQL query (use triple quotes for multi-line queries)
query = """
  select * from crypto2.daily_bitcoin_summary order by transaction_date;
"""

# 3. Run the query and convert directly to a Pandas DataFrame
query_job = client.query(query)
df = query_job.to_dataframe()

# 4. Use your data
print(df.head())
```
